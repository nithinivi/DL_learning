## Beale Fucntion 
[https://en.wikipedia.org/w/index.php?title=Test_functions_for_optimization&action]

$f(x,y) = \left( 1.5 - x + xy \right)^{2} + \left( 2.25 - x + xy^{2}\right)^{2} + \left(2.625 - x+ xy^{3}\right)^{2}$

```python
f  = lambda x , y : (1.5 - x + x*y)**2 + (2.25 - x + x*y**2) + (2.625 - x + x*y**3 )**2
```

```python
# Ranges 
xmin , xmax , xstep  = -4.5 , 4.5 , .2
ymin , ymax , ystep  = -4.5 , 4.5 , .2
```

```python
x,y = np.meshgrid(np.arange(xmin, xmax + xstep, xstep),
                 np.arange(ymin, ymax + ystep, ystep))
```

```python
z = f(x,y)
```

```python
minima = np.array([3., .5])

f(*minima) # the global minium of the "f"
```

```python
minima_ = minima.reshape(-1, 1)
minima_ ,f(*minima_) 
```

### 3D Sufrace Plot

```python
fig = plt.figure(figsize=(8,5))
ax = plt.axes(projection="3d", elev=50, azim=-50)  # the direction in which u view the plot
ax.plot_surface(x,y,z,norm=LogNorm(),  # normalizing the color
                rstride=1,cstride=1,   # 
                alpha=.8,edgecolor=None,
                cmap= plt.cm.jet)

ax.plot(*minima_, f(*minima_), 'r*', markersize=10) #minima

ax.set_xlabel('$x$') #latex
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')

ax.set_xlim((xmin, xmax))
ax.set_ylim((ymin, ymax))

plt.show()
```

# Himmelblau's function

$f(x,y)=(x^{2}+y-11)^{2}+(x+y^{2}-7)^{2}.\quad $

```python
f1  = lambda x , y : (x**2 + y - 11)**2 + (x + y**2 -7)**2
```

```python
# Ranges 
xmin , xmax , xstep  = -10 , 10 , .2
ymin , ymax , ystep  = -10 , 10 , .2
```

```python
x,y = np.meshgrid(np.arange(xmin, xmax + xstep, xstep),
                 np.arange(ymin, ymax + ystep, ystep))
z = f1(x,y)
```

```python
minima = np.array([3., 2])
f1(*minima) # the global minium of the "f"
```

```python
minima_ = minima.reshape(-1, 1)
minima_ ,f1(*minima_) 
```

```python
fig = plt.figure(figsize=(8,5))
ax = plt.axes(projection="3d", elev=60, azim=-50)  # the direction in which u view the plot
ax.plot_surface(x,y,z,norm=LogNorm(),# normalizing the color
                rstride=1,cstride=1,   # 
                alpha=.8,edgecolor=None,
                cmap= plt.cm.jet)

ax.plot(*minima_, f1(*minima_), 'g*', markersize=10) #minima

ax.set_xlabel('$x$') #latex
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')

ax.set_xlim((xmin, xmax))
ax.set_ylim((ymin, ymax))

plt.show()
```
